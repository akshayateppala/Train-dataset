```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, f_classif

# -------------------------------------
# Load Dataset
# -------------------------------------

df = pd.read_csv("train.csv")

print("First 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nDataset Information")
print(df.info())
```

```python
# ==================================================
# 1. DATA CLEANING
# ==================================================

print("\nMissing Values")
print(df.isnull().sum())

# -------------------------------------
# Fill missing Age with Mean
# -------------------------------------

age_imputer = SimpleImputer(strategy='mean')
df['Age'] = age_imputer.fit_transform(df[['Age']])

# Fill missing Embarked with Mode

embarked_imputer = SimpleImputer(strategy='most_frequent')
df['Embarked'] = embarked_imputer.fit_transform(df[['Embarked']]).ravel()

# Drop Cabin because too many missing values

df.drop(columns=['Cabin'], inplace=True)

print("\nMissing Values After Cleaning")
print(df.isnull().sum())
```

```python
# ==================================================
# 2. REMOVE DUPLICATES
# ==================================================

print("\nDuplicate Rows:", df.duplicated().sum())

df.drop_duplicates(inplace=True)
```

```python
# ==================================================
# 3. OUTLIER DETECTION (Age)
# ==================================================

Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Age'] < lower) | (df['Age'] > upper)]

print("\nAge Outliers")
print(outliers[['PassengerId','Age']])
```

```python


# Import Libraries
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold

# -------------------------------
# Step 1: Load Dataset
# -------------------------------
df = pd.read_csv("train.csv")

print("Original Dataset")
print(df.head())

# -------------------------------
# Step 2: Data Cleaning
# -------------------------------

# Remove duplicate rows
df = df.drop_duplicates()

# Fill missing values
for column in df.select_dtypes(include=['number']).columns:
    df[column] = df[column].fillna(df[column].mean())

for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].fillna(df[column].mode()[0])

print("\nAfter Cleaning")
print(df.head())

# -------------------------------
# Step 3: Data Integration
# -------------------------------
# (Used when combining multiple datasets)

# Example:
# df2 = pd.read_csv("dataset2.csv")
# df = pd.merge(df, df2, on="ID")

print("\nData Integration Step Completed")

# -------------------------------
# Step 4: Data Transformation
# -------------------------------

# Normalize numerical columns
scaler = MinMaxScaler()

num_cols = df.select_dtypes(include=['number']).columns

df[num_cols] = scaler.fit_transform(df[num_cols])

print("\nAfter Transformation")
print(df.head())

# -------------------------------
# Step 5: Data Reduction
# -------------------------------

selector = VarianceThreshold(threshold=0.01)

reduced_data = selector.fit_transform(df[num_cols])

print("\nShape Before Reduction:", df[num_cols].shape)
print("Shape After Reduction :", reduced_data.shape)

# -------------------------------
# Final Dataset
# -------------------------------
print("\nData Preprocessing Completed Successfully!")
```

```python
# ==================================================
# IMPORT LIBRARIES
# ==================================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    LabelEncoder
)

from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.decomposition import PCA

from scipy.stats import zscore


# ==================================================
# LOAD TITANIC DATASET
# ==================================================

df = pd.read_csv("train.csv")

print("Initial Dataset")
print(df.head())

print("\nShape:")
print(df.shape)


# ==================================================
# 1. DATA CLEANING
# ==================================================

# -------------------------------
# Missing Value Detection
# -------------------------------

print("\nMissing Values")
print(df.isnull().sum())


# -------------------------------
# Mean Imputation (Age)
# -------------------------------

age_imputer = SimpleImputer(strategy="mean")

df["Age"] = age_imputer.fit_transform(
    df[["Age"]]
)


# -------------------------------
# Mode Imputation (Embarked)
# -------------------------------

mode_imputer = SimpleImputer(
    strategy="most_frequent"
)

df["Embarked"] = mode_imputer.fit_transform(
    df[["Embarked"]]
).ravel()


# -------------------------------
# Drop Cabin column
# -------------------------------

df.drop(
    columns=["Cabin"],
    inplace=True
)


print("\nAfter Missing Value Handling")
print(df.isnull().sum())


# ==================================================
# Predictive Imputation using KNN
# ==================================================

knn = KNNImputer(
    n_neighbors=3
)

df[["Age","Fare"]] = knn.fit_transform(
    df[["Age","Fare"]]
)


# ==================================================
# Remove Duplicate Data
# ==================================================

print("\nDuplicate Rows:")
print(df.duplicated().sum())

df.drop_duplicates(
    inplace=True
)


# ==================================================
# Noise Handling - Binning
# ==================================================

df["Fare_Bin"] = pd.qcut(
    df["Fare"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)

print("\nFare Binning")
print(df[["Fare","Fare_Bin"]].head())


# ==================================================
# Outlier Detection
# ==================================================

z_scores = zscore(df["Age"])

outliers = df[
    abs(z_scores) > 3
]

print("\nAge Outliers")
print(outliers)


# Visualization

plt.figure(figsize=(6,4))

sns.boxplot(
    x=df["Age"]
)

plt.title(
    "Age Outlier Detection"
)

plt.show()



# ==================================================
# 2. DATA INTEGRATION
# ==================================================

# Example: Split data source

customer = df[
    ["PassengerId","Name"]
]

ticket = df[
    ["PassengerId","Ticket"]
]


merged_data = pd.merge(
    customer,
    ticket,
    on="PassengerId"
)


print("\nIntegrated Dataset")
print(merged_data.head())



# ==================================================
# 3. DATA TRANSFORMATION
# ==================================================


# -------------------------------
# Label Encoding
# -------------------------------

encoder = LabelEncoder()

df["Sex"] = encoder.fit_transform(
    df["Sex"]
)

# male = 1
# female = 0



# -------------------------------
# One Hot Encoding
# -------------------------------

df = pd.get_dummies(
    df,
    columns=["Embarked"],
    drop_first=True
)



# -------------------------------
# Normalization
# -------------------------------

scaler = MinMaxScaler()

df[
    ["Age","Fare"]
] = scaler.fit_transform(
    df[
        ["Age","Fare"]
    ]
)


print("\nNormalized Data")
print(df[["Age","Fare"]].head())



# -------------------------------
# Standardization
# -------------------------------

standard = StandardScaler()


df[
    ["Age","Fare"]
] = standard.fit_transform(
    df[
        ["Age","Fare"]
    ]
)


print("\nStandardized Data")
print(df[["Age","Fare"]].head())



# -------------------------------
# Aggregation
# -------------------------------

print("\nAverage Fare by Class")

print(
    df.groupby(
        "Pclass"
    )["Fare"].mean()
)



# -------------------------------
# Discretization
# -------------------------------

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=4,
    labels=[
        "Child",
        "Young",
        "Adult",
        "Senior"
    ]
)


print("\nAge Groups")
print(
    df[
        ["Age","Age_Group"]
    ].head()
)



# ==================================================
# 4. DATA REDUCTION
# ==================================================


# Remove unnecessary columns

df.drop(
    columns=[
        "PassengerId",
        "Name",
        "Ticket",
        "Fare_Bin"
    ],
    inplace=True
)



# Convert categories

df = pd.get_dummies(df)


# Separate input and output

X = df.drop(
    "Survived",
    axis=1
)

y = df["Survived"]



# ==================================================
# Feature Selection
# ==================================================

selector = SelectKBest(
    score_func=f_classif,
    k=5
)


X_selected = selector.fit_transform(
    X,
    y
)


print("\nFeature Selection Result")

print(
    X_selected.shape
)



# Selected feature names

selected_features = X.columns[
    selector.get_support()
]


print(
    "Selected Features:"
)

print(
    selected_features
)



# ==================================================
# PCA Dimensionality Reduction
# ==================================================

pca = PCA(
    n_components=2
)


X_pca = pca.fit_transform(
    X_selected
)


print("\nPCA Result")

print(
    X_pca.shape
)



# ==================================================
# Numerosity Reduction - Sampling
# ==================================================

sample_data = df.sample(
    n=50,
    random_state=42
)


print("\nSample Data")

print(
    sample_data.head()
)



# ==================================================
# FINAL PREPROCESSED DATA
# ==================================================

print("\nFinal Dataset")

print(
    df.head()
)

print("\nFinal Shape")

print(
    df.shape
)
```
